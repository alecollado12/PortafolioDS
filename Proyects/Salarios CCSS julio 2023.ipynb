## CCSS Salaries for July 2023

You can find the updated data at the following link: https://www.ccss.sa.cr/datos-abiertos-salarios

The salaries are expressed in gross amounts, meaning they are before taxes and deductions.

The purpose of this notebook is to analyze how CCSS salaries are distributed by position and what salary ranges a new employee can expect.

The notebook consists of two parts:

1) Data transformation and cleaning.

2) Data analysis.

## Data Cleaning and Transformation

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
# Loadding the dataset
df = pd.read_csv('C:/Users/Usuario/Documents/PortafolioDS/DATA/ccss_salarios.csv', sep=';')
df
```

```python
# We set the style for the graphs.
sns.set(style = 'darkgrid', context = 'notebook')
```

```python
df.dtypes
```

```python
# change the column names.
df.columns = ['Id','nombre','puesto','salario_colones']

# remove columns that are not of interest.
df = df.drop('Id', axis = 1)

# change the data type to string.
df['nombre'] = df['nombre'].astype('string')
df['puesto'] = df['puesto'].astype('string')

# remove the colón symbols and eliminate characters that come after the ','.
df['salario_colones'] = df['salario_colones'].str.replace('₡','')
df['salario_colones'] = df['salario_colones'].str.replace('.','').str.split(',').str.get(0).astype('float')

# set the pandas format option to use commas as thousands separators.
pd.options.display.float_format = '{:,.0f}'.format
```

```python
df.dtypes
```

```python
# how many null values we have
df.isnull().sum()
```

```python
# remove null values
df = df.dropna()
df
```

```python
# number of employes that report 0 as salary
df[df['salario_colones'] == 0]['salario_colones'].count()
```

```python
# remove values = 0
df.drop(df[df['salario_colones'] == 0].index, inplace = True)
df = df.reset_index(drop = True)
df
```

```python
# How many positions do we have?
df['puesto'].nunique()
```

```python
# Some of the positions are:
df['puesto'].unique()
```

Analyzing 429 different positions would be a very tedious task, so we should improve the grouping of these positions. This way, we can analyze by category and job specialization or title.

```python
# We create two new columns based on the job position.
df[['categoria','titulo']] = df['puesto'].str.split(' ',n=1,expand = True)
df
```

```python
df['categoria'].unique()
```

We've already reduced the number of positions by half, but since they are separated by '.' and spaces, we can apply the same method again to see if we can reduce them further.

```python
df['categoria'] = df['categoria'].str.split('.', n=1).str.get(0)
df
```

```python
# Let's review the number of categories again.
print(df['categoria'].nunique())
df['categoria'].unique()
```

We have successfully reduced/grouped the positions into 92 categories by splitting on spaces and commas. However, we still see that there are records that mean the same thing but are written differently, such as 'AUX' and 'AUXILIAR'. Therefore, we can further reduce the number of unique values in our category column.

```python
# We create a replacement dictionary with their respective keys and values.
reemplazos = {
    'AUX': 'AUXILIAR',
    'TRAB': 'TRABAJADOR',
    'TEC': 'TECNICO',
    'JEF': 'JEFE',
    'SUP': 'SUPERIVISOR',
    'OFI': 'OFICINISTA',
    'GEREN':'GERENTE',
    'ASIS': 'ASISTENTE',
    'ANAL': 'ANALISTA',
    'OPER': 'OPERADOR',
    'SEC': 'SECRETARIA',
    'ASES': 'ASESOR',
    'ENF': 'ENFERMERA',
    'MED': 'MEDICO',
    'DIR': 'DIRECTOR',
    'FARMA': 'FARMACEUTICO',
    'DIPL': 'DIPLOMADO',
    'ODONT': 'ODONTOLOGO',
    'BODE': 'BODEGUERO',
    'ADM':'ADMINISTRADOR'
}

# Function to replace values
def aplicar_reemplazos(valor):
    for original, reemplazo in reemplazos.items():
        if valor.startswith(original) and valor != reemplazo:
            return reemplazo
    return valor

# We apply the replacement function to the 'category' column.
df['categoria'] = df['categoria'].apply(aplicar_reemplazos)
df['categoria'] = df['categoria'].replace('^M$','MAE',regex = True)
df['categoria'] = df['categoria'].replace('^AS$','ASISTENTE',regex = True)
df['categoria'] = df['categoria'].replace('^PROF$','PROFESIONAL',regex = True)
print(df['categoria'].nunique())
df['categoria'].unique()
```

The 'category' column is already reduced to its maximum grouping. Now we should complete the same process with the 'title' column.

Note: In the case of the 'MAE' category, even though it appears to belong to the medical category based on the description provided in the 'position' column, we will leave it as an independent category since that is how it was classified when registered in the CCSS database.

```python
# We create another column, but this time splitting by '.' instead of a space.
df['titulo2'] = df['puesto'].str.split('.', n=1, expand = True).get(1)
df
```

```python
# We add the values from 'title2' to 'title'. 
df['titulo'] = df.apply(lambda row: row['titulo2'] if pd.isnull(row['titulo']) and not pd.isnull(row['titulo2']) else row['titulo'], axis=1)
df
```

```python
# remove the column 'titulo2'
df = df.drop('titulo2', axis = 1)
```

```python
df
```

## Analysis

Now that our data is clean and enriched, we can proceed with the analysis.

```python
df.describe()
```

```python
plt.figure(figsize=(15,4))

sns.boxplot(data=df,x='salario_colones')

formateador = ticker.StrMethodFormatter("{x:,.0f}")
plt.gca().xaxis.set_major_formatter(formateador)
```

Let's start by determining the total salary range.

As we can see in the previous boxplot, the reported salary for the lowest-paid CCSS employee is very close to 0, precisely 37 colones. Meanwhile, the salary of the highest-paid individual is slightly above 17.5 million colones.

Let's exclude the extreme values and see how much most CCSS employees earn.

```python
plt.figure(figsize=(15,4))

sns.boxplot(data=df[df['salario_colones'] < df['salario_colones'].quantile(0.95)],x='salario_colones')

formateador = ticker.StrMethodFormatter("{x:,.0f}")
plt.gca().xaxis.set_major_formatter(formateador)
```

### Now let's look at it by category.

```python
plt.figure(figsize=(15,70))

sns.boxplot(data=df[df['salario_colones'] < 7500000],x='salario_colones',y='categoria')

formateador = ticker.StrMethodFormatter("{x:,.0f}")
plt.gca().xaxis.set_major_formatter(formateador)
```

```python
# We sort by category to find the one with the most employees and its average income.
cuenta = df.groupby('categoria')['salario_colones'].agg(Ingreso_promedio='mean', recuento_empleados='count').sort_values('recuento_empleados', ascending = False)
cuenta
```

#### To analyze each category in more detail with its titles, we can filter the dataframe.

```python
# Lets create a filtered dataframe for the category we want to analyze in depth.
categoria_filtrar = 'AUXILIAR'

df_filtrado = df[df['categoria'] == categoria_filtrar]
df_filtrado = df_filtrado.reset_index(drop = True)
df_filtrado
```

```python
titulos = df_filtrado.groupby('titulo')['salario_colones'].agg(Ingreso_promedio='mean', recuento_empleados='count').sort_values('Ingreso_promedio', ascending = False)
titulos
```

```python
# size of the chart
plt.figure(figsize=(15,10))

sns.stripplot(data = df_filtrado,y = 'categoria', x = 'salario_colones', hue = 'titulo')

# Configure the thousands separators on the x-axis.
formateador = ticker.StrMethodFormatter("{x:,.0f}")
plt.gca().xaxis.set_major_formatter(formateador)

# We remove the legend from the chart
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# size of the chart
canti_tit  = df_filtrado['titulo'].nunique()*1.5

plt.figure(figsize=(15,canti_tit))

sns.boxplot(data = df_filtrado,y = 'titulo', x = 'salario_colones')

# Configure the thousands separators on the x-axis.
formateador = ticker.StrMethodFormatter("{x:,.0f}")
plt.gca().xaxis.set_major_formatter(formateador)
plt.title(f'Rangos salariales de {categoria_filtrar} segun su especialidad')
```
